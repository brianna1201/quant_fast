### 객체 메서드

```python
class Position:
    def __init__(self, shrs=0, px=0):
        self.shares = shrs
        self.price = px

    def mktValue(self):
        return self.shares * self.price
```

```python
pos = Position(100, 50)
```

```python
mktval = pos.mktValue()
```

### 파이썬 클래스를 활용하여 오징어게임 구현하기

```python
# Numpy 패키지 임포트
import numpy as np
```

```python
# SquidGame 클래스 생성
class SquidGame:

    # 생성자(__init__) 함수
    # 입력변수: 참가번호(number)
    def __init__(self, number):

        # 참가번호
        self.number = number

        # 현재 상태
        self.status = '생존'


    # 첫번째 게임 플레이 함수
    def play_first_game(self):

        # 만약 상태가 생존이라면, 랜덤 넘버를 하나 뽑기
        if self.status == '생존':

            # 랜덤 넘버(random_num)
            random_num = np.random.rand(1)

            # 랜덤 넘버가 0.5보다 크다면
            if random_num >= 0.5:

                # 게임에서 탈락, 상태가 사망으로 바뀜
                self.status = '사망'

                # 탈락 메시지 출력
                print(str(self.number) + '번, 첫번째 게임 탈락')

            # 그렇지 않다면
            else:
                # 통과 메시지 출력
                print(str(self.number) + '번, 첫번째 게임 통과')

        # 상태가 생존이 아니라면, 이미 탈락 메시지 출력
        else:
            print('당신은 이미 탈락하셨습니다. 게임을 플레이할 수 없습니다.')

    # 두번째 게임 플레이 함수
    def play_second_game(self):

        # 만약 상태가 생존이라면, 랜덤 넘버를 하나 뽑기
        if self.status == '생존':

            # 랜덤 넘버(random_num)\
            random_num = np.random.rand(1)

            # 랜덤 넘버가 0.25보다 크다면
            if random_num >= 0.25:

                # 게임에서 탈락, 상태가 사망으로 바뀜
                self.status == '사망'

                # 탈락 메시지 출력
                print(str(self.number) + '번, 두번째 게임 탈락')

            # 그렇지 않다면
            else:
                # 통과 메시지 출력
                print(str(self.number) + '번, 두번째 게임 통과')

        # 상태가 생존이 아니라면, 이미 탈락 메시지 출력
        else:
            print('당신은 이미 탈락하셨습니다. 게임을 플레이할 수 없습니다.')

    # 세번째 게임 플레이 함수
    def play_third_game(self):

        # 만약 상태가 생존이라면, 랜덤 넘버를 하나 뽑기
        if self.status == '생존':

            # 랜덤 넘버(random_num)
            random_num = np.random.rand(1)

            # 랜덤 넘버가 0.125보다 크다면
            if random_num >= 0.125:

                # 게임에서 탈락, 상태가 사망으로 바뀜
                self.status = '사망'

                # 탈락 메시지 출력
                print(str(self.number) + '번, 세번째 게임 탈락')

            # 그렇지 않다면
            else:
                # 통과 메시지 출력
                print(str(self.number) + '번, 세번째 게임 통과')

        # 상태가 생존이 아니라면, 이미 탈락 메시지 출력
        else:
            print('당신은 이미 탈락하셨습니다. 게임을 플레이할 수 없습니다.')
```
